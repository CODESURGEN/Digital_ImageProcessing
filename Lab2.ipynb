```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
n = 5
```

```python
gray_img =  cv2.imread("lena.png",0) 
plt.imshow(gray_img, cmap='gray')
plt.show()

# reading lena image into gray scale 
```

```python
noise_store = np.zeros((n,gray_img.shape[0],gray_img.shape[1]) ,dtype = np.uint8)
# creating an empty array to store the noise images 
```

```python
for i in range(n):
    noise =np.random.randn(*gray_img.shape)*50
    #noise = (noise -noise.mean())/noise.std()
    noise_store[i] = np.clip(gray_img + noise, 0,255)
```

```python
  

for i in range(n):
    plt.imshow(noise_store[i], cmap='gray')
    plt.title("Noisy Image " + str(i + 1) , color = "white")
    plt.xticks([]), plt.yticks([])

    plt.show()
```

```python
# Find the average image
g = np.mean(noise_store , axis=0).astype(np.uint8)
# Display the average image
plt.imshow(g, cmap='gray')
plt.show()
```
