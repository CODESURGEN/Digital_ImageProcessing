```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
```

```python
gray_img =  cv2.imread("lena.png",0) 
plt.imshow(gray_img, cmap='gray')
plt.show()
```

```python
gray_img.shape
```

```python
resize = [(128,128),(64,64),(32,32),(16,16)]
```

```python

for size in resize:
    resized_gray = cv2.resize(gray_img, size, interpolation=cv2.INTER_LINEAR)
    resized_gray = cv2.resize(resized_gray, gray_img.shape[::-1], interpolation=cv2.INTER_LINEAR)
    vis = np.concatenate((gray_img, resized_gray), axis=1)

    plt.imshow(vis, cmap='gray')
    plt.title("lena resized with size" + str(size),  color = "white")
    plt.xticks([]), plt.yticks([])

    plt.show()

```

```python

# Intensity ranges
intensity_ranges = [255, 128, 64, 32, 16, 8, 4, 2]

# Downsample image for each intensity range
downsampled_images = []
for r in intensity_ranges:
    downsampled_image = (gray_img // (255/r))
    downsampled_images.append(downsampled_image)

# Display all 8 downsampled images
fig, axes = plt.subplots(2, 4, figsize=(14, 8))
ax = axes.ravel()
for i, image in enumerate(downsampled_images):
    ax[i].imshow(image, cmap='gray')
    ax[i].axis('off')
    ax[i].set_title(f'Intensity range: 0-{intensity_ranges[i]}')
plt.show()
```
